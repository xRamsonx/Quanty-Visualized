```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-colorblind')
```

```python
def get_maxima_in_range(xas,energy_range):
    x=xas[0]
    idx=(x>energy_range[0]) & (x<energy_range[1])
    #idx=np.where(x>energy_range[0] & x<energy_range[1])
    x=x[idx]
    y=-xas[2][idx]
    x_max=x[np.where(y==max(y))]
    return x_max[0]

def get_M4(xas):
    return get_maxima_in_range(xas,[0,1000])

def get_M5(xas):
    return get_maxima_in_range(xas,[-1000,0])

def print_edges(xas):
    print("M4:",get_M4(xas),"eV")
    print("M5:",get_M5(xas),"eV")
    
def get_XAS_range(xas1,xas2,delta1,delta2):
    M4=np.mean([get_M4(xas1),get_M4(xas2)])
    M5=np.mean([get_M5(xas1),get_M5(xas2)])
    return [M4-delta1,M4+delta2], [M5-delta1,M5+delta2]
    
```

# UF6

```python
large_basis = np.loadtxt("UF6/data/XAS/large_basis",skiprows=5).T
small_basis = np.loadtxt("UF6/data/XAS/small_basis",skiprows=5).T

fig =plt.figure(figsize=(12,6))
ax1 = plt.subplot(121)
ax1.plot(small_basis[0],-small_basis[2],label="small")
ax1.plot(large_basis[0],-large_basis[2],label="large")
ax1.set_title("M5")
ax1.legend()
ax1.set_xlim(get_XAS_range(small_basis,large_basis,8,12)[0])

ax2 = plt.subplot(122)
ax2.plot(small_basis[0],-small_basis[2],label="small")
ax2.plot(large_basis[0],-large_basis[2],label="large")
ax2.set_title("M4")
ax2.legend()
ax2.set_xlim(get_XAS_range(small_basis,large_basis,8,12)[1])
```
