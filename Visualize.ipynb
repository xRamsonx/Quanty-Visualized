```python
def shift_to_experiment(experiment,xas,energy_range):
    experiment_max=experiment[0][np.where(experiment[1]==max(experiment[1]))]
    x=xas[0]
    idx=(x>energy_range[0]) & (x<energy_range[1])
    #idx=np.where(x>energy_range[0] & x<energy_range[1])
    x=x[idx]
    y=xas[1][idx]
    x_max=x[np.where(y==max(y))]
    return (x-x_max)/1000+experiment_max,y
```

```python
large_basis = np.loadtxt("Ba3UO6/data/Matrix/large_basis",skiprows=5).T
small_basis = np.loadtxt("Ba3UO6/data/Matrix/small_basis",skiprows=5).T

x_large = large_basis[0]
y_large = -large_basis[2]
y_large = y_large/max(y_large)
x_small = small_basis[0]
y_small = -small_basis[2]
y_small = y_small/max(y_small)

x_large, y_large = shift_to_experiment((x_experiment,y_experiment),(x_large,y_large),(-565,-540))
x_small, y_small = shift_to_experiment((x_experiment,y_experiment),(x_small,y_small),(-565,-540))
plt.plot(x_large+0.00005, y_large,label="with d")#1.6001
plt.plot(x_small+0.00005, y_small,label="without d")#1.491
plt.legend()
#plt.plot(x_experiment,y_experiment)
#plt.plot(small_basis2[0],-small_basis2[2])#1.491
plt.xlim(3.542,3.568)
plt.show()
```

```python
def getRixs(path):
    RIXS=np.loadtxt(path,skiprows=5)
    RIXS = RIXS.T[1:]
    #RIXS_Real=[]
    RIXS_Img=[]
    for i in range(0,int(len(RIXS)/2)):
        #RIXS_Real.append(RIXS[2*i])
        RIXS_Img.append(RIXS[2*i+1])
    return -np.array(RIXS_Img).T
xmin=805
xmax=828
ymin=-13.75
ymax=28.75

RIXS_small=getRixs("Ba3UO6/data/RIXS/No/M4/RIXS_z_z.dat")
RIXS_large=getRixs("Ba3UO6/data/RIXS/6d/M4/RIXS_z_z.dat")
shift = RIXS_small.T[6].argmax()-RIXS_large.T[6].argmax()
print(shift)
#print(np.sum(RIXS_small.T[6]),np.sum(RIXS_large.T[6]))
#print(np.sum(RIXS_small),np.sum(RIXS_large))
#print(RIXS_small[:][328].argmax())
RIXS_small=RIXS_small[7000:]
RIXS_large=RIXS_large[7000-shift:-shift]

ax1 = plt.subplot(121)
ax1.set_title("without 6d")
ax1.set_ylabel("E2")
ax1.set_xlabel("E1")
ax1.imshow(RIXS_small/RIXS_small.max(),extent=[xmin,xmax,ymin,ymax])

ax2 = plt.subplot(122)
ax2.set_title("with 6d")
ax2.set_xlabel("E1")
ax2.axes.get_yaxis().set_visible(False)
im = ax2.imshow(RIXS_large/RIXS_large.max(),extent=[xmin,xmax,ymin,ymax])
divider = make_axes_locatable(ax2)
cax= divider.append_axes("right",size="5%",pad=0.05)
plt.colorbar(im, cax=cax, orientation="vertical")
plt.show()

diff=RIXS_small/RIXS_small.max()-RIXS_large/RIXS_large.max()
plt.title("Difference")
plt.ylabel("E2")
plt.xlabel("E1")
plt.imshow(diff,extent=[xmin,xmax,ymin,ymax],aspect=.6)
plt.colorbar()
plt.show()
```
